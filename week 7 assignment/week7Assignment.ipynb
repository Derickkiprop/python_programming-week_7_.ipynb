```python
import pandas as pd

# Define file path
dataset_path = 'iris.csv'

try:
    # Attempt to read the CSV file
    df = pd.read_csv(dataset_path)
    print("File loaded successfully!")
except FileNotFoundError:
    print(f"Error: The file at {dataset_path} was not found.")
except pd.errors.ParserError:
    print("Error: There was an issue with parsing the file. Ensure the file is in CSV format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import pandas as pd  # Ensure pandas is imported
from sklearn.datasets import load_iris  # Load Iris from sklearn
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df['species'] = iris.target
```

```python
print(df.head())
```

```python
try:
    # Check for missing data
    missing_data = df.isnull().sum()
    if missing_data.any():
        print("Missing data found. Handling missing values...")
        # Fill missing values with the mean of each column
        df.fillna(df.mean(), inplace=True)
        print("Missing values filled with column means.")
    else:
        print("No missing data found.")
except AttributeError:
    print("Error: The dataset might not have been loaded correctly.")
```

```python
print(pd.__version__)
```

```python
try:
    # Try to convert a column to a numeric type
    df['sepal length (cm)'] = pd.to_numeric(df['sepal length (cm)'], errors='raise')
except ValueError as e:
    print(f"Error: There is an issue with the data in 'sepal length (cm)': {e}")
except KeyError:
    print("Error: The column 'sepal length (cm)' does not exist in the dataset.")
except Exception as e:
    print(f"An unexpected error occurred while processing data types: {e}")
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target  # Add the species column
```

```python
print(df.head())
```

```python
# Bar chart: Average petal length per species
species_avg_petal = df.groupby("species")["petal length (cm)"].mean()

plt.figure(figsize=(8, 5))
species_avg_petal.plot(kind="bar", color=["green", "orange", "blue"])
plt.title("Average Petal Length by Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Average Petal Length (cm)", fontsize=12)
plt.xticks([0, 1, 2], labels=["Setosa", "Versicolor", "Virginica"], rotation=0)  # Customize species labels
plt.show()
```

```python
# Histogram: Distribution of sepal length
plt.figure(figsize=(8, 5))
plt.hist(df["sepal length (cm)"], bins=10, color="purple", edgecolor="black")
plt.title("Distribution of Sepal Length", fontsize=16)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
# Scatter plot: Sepal length vs. Petal length
plt.figure(figsize=(8, 5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df, palette="bright")
plt.title("Sepal Length vs. Petal Length", fontsize=16)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)
plt.legend(title="Species", labels=["Setosa", "Versicolor", "Virginica"])
plt.show()
```
